```python
import cv2
import time
import mediapipe as mp
```

```python
video = cv2.VideoCapture(0)

previous_time, new_time = (0, 0)

mpHands = mp.solutions.hands
hands = mpHands.Hands()

drawing_hands = mp.solutions.drawing_utils

while True:
    retval, img = video.read()
    img = cv2.flip(src=img, flipCode=1)
    
    
    # ----------------   FPS ----------------------------
    new_time = time.time()
    fps = 1 / (new_time - previous_time)
    previous_time = new_time
    
    img = cv2.putText(img=img,
                      text=f'FPS: {int(fps)}',
                      org=(30, 40),
                      fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                      fontScale=1,
                      thickness=4,
                      color=(0, 0, 255))
     #! ----------------   FPS ----------------------------
        
        
    landmarkList = []
    
    # ----------------------- HAND DETECTION ---------------------------------------
    results = hands.process(image=img).multi_hand_landmarks
    
    if results:
        for landmarks in results:
                drawing_hands.draw_landmarks(image=img,
                                             landmark_list=landmarks,
                                             connections=mpHands.HAND_CONNECTIONS)
                
                for id, landmark in enumerate(landmarks.landmark):
                    height, width, color = img.shape
                    cx, cy = int(landmark.x * width), int(landmark.y * height)
                    landmarkList.append([id, cx, cy])
    #! ----------------------- HAND DETECTION ---------------------------------------
                    
    
    # -------------------------- LEFT - RIGHT HAND DETECTION ------------------------
        if len(landmarkList) != 0:
            
            if landmarkList[4][1] < landmarkList[8][1]:
                cv2.putText(img=img,
                            text=f"Right Hand",
                            org=(30, 70),
                            fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                            fontScale=1,
                            color=(0, 0, 255),
                            thickness=4)
            else:
                cv2.putText(img=img,
                            text="Left Hand",
                            org=(30, 70),
                            fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                            fontScale=1,
                            color=(0, 0, 255),
                            thickness=4)
    #! -------------------------- LEFT - RIGHT HAND DETECTION ------------------------
    
    cv2.imshow(winname='Video', mat=img)
    if cv2.waitKey(1) == ord('q'): break
video.release()
cv2.destroyAllWindows()
```
